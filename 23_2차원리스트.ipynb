```python
# 반복문으로 2차원 리스트의 요소를 모두 출력하기 
a = [
    [10, 20], 
    [30, 40], 
    [50, 60]
]

for x, y in a:
    print(x, y)
```

```python
# for 반복문을 두 번 사용하기 
a = [
    [10, 20], 
    [30, 40], 
    [50, 60]
]

for i in a:
    for j in i:
        print(j, end=" ")

    print()
```

```python
a = [
    [10, 20], 
    [30, 40], 
    [50, 60]
]

for i in range(len(a)):
    for j in range(len(a[i])):
        print(a[i][j], end=" ")
    print()
```

```python
# while 반복문을 한 번 사용하기 
a = [
    [10, 20], 
    [30, 40], 
    [50, 60]
]

i = 0
while i < len(a):
    x, y = a[i] 
    print(x, y)
    i += 1
```

```python
# while 반복문을 두 번 사용하기 
a = [
    [10, 20], 
    [30, 40], 
    [50, 60]
]

i = 0
while i < len(a): 
    j = 0 
    while j < len(a[i]):
        print(a[i][j], end=" ")
        j += 1
    print()
    i += 1 
```

```python
# 톱니형 리스트 만들기
a = [3, 1, 3, 2, 5]
b = []

for i in a:
    line = [] 
    for _ in range(i):
        line.append(0)
    b.append(line)


print(b)
```

```python
row, col = map(int, input().split())

matrix = [] 
for i in range(row):
    matrix.append(list(input())) 
```

```python
matrix
```

```python
for r in range(row):
    for c in range(col): 
        if matrix[r][c] == "*":
            continue

        count = 0 
        for nr in range(r-1, r+2): 
            for nc in range(c-1, c+2): 

                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == "*":
                        count += 1
        matrix[r][c] = str(count)

for r in range(row): 
    print("".join(matrix[r]))
```
