```python
# 위치 인수와 리스트 언패킹 
## 위치인수를 사용하는 함수

def print_numbers(a, b, c): 
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30)
```

```python
## 언패킹 사용하기 

x = [10, 20, 30]
print(x)
print(*x)
# *[10, 20, 30]
print(10, 20, 30)
```

```python
print_numbers(*x)
# print_numbers(10, 20, 30)
```

```python
def print_numbers(a, b, c): 
    print(a)
    print(b)
    print(c)


print_numbers(10, 20)
```

```python
## 가변 인수 함수 만들기

def print_numbers(*args):
    print(args)
    for arg in args:
        print(arg)
```

```python
print_numbers(10)
```

```python
print_numbers(10, 20, 30, 40)
```

```python
x = [10]
y = [10, 20, 30, 40]
10, 20, 30, 40

print_numbers(*x)
print_numbers(*y)
```

```python
# 키워드 인수 사용하기 
def personal_info(name, age, address): 
    print("이름 : ", name)
    print("나이 : ", age)
    print("주소 : ", address)
```

```python
personal_info('홍길동', 30, '서울시 용산구')
```

```python
personal_info(name='홍길동', age=30, address='서울시 용산구')
```

```python
personal_info(age=30, address="서울시 용산구", name="홍길동")
```

```python
personal_info(age=30, name="홍길동", "서울시 용산구")
```

```python
# 키워드 인수와 딕셔너리 언패킹 사용하기
def personal_info(name, age, address): 
    print("이름 : ", name)
    print("나이 : ", age)
    print("주소 : ", address)
```

```python
x = {
    'name': '홍길동', 
    'age': 30, 
    'address': '서울시 용산구'
}


personal_info(**x)
# personal_info(name='홍길동', age=30, address='서울시 용산구')
```

```python
# 키워드 인수를 사용하는 가변 인수 함수 만들기 

def personal_info(**kwargs):
    # print(kwargs)
    # print(type(kwargs))
    for kw, arg in kwargs.items():
        print(kw, ": ", arg, sep="")
```

```python
personal_info(name="홍길동", age=30)
```

```python
personal_info(name="홍길동", age=30, address="서울시 용산구")
```

```python
x = {
    "name" : "홍길동"
}

y = {
    'name': '홍길동', 
    'age': 30, 
    'address': '서울시 용산구'
}
```

```python
personal_info(**x)
personal_info(**y)
```
