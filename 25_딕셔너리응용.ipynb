```python
## 딕셔너리 표현식 사용하기 

# {키:값 for 키, 값 in 딕셔너리 }
# dict({키:값 for 키, 값 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']
imsi = {
    'a' : None, 
    'b' : None, 
    'c' : None, 
    'd' : None
}


# {for key, value in dict.fromkeys(keys).items()}
x = {a: b for a, b in imsi.items()}
x
```

```python
{key : 0 for key in dict.fromkeys(['a', 'b', 'c', 'd']).keys()}
```

```python
{value : 0 for value in {
    "a" : 10, 
    "b" : 20, 
    "c" : 30, 
    "d" : 40
}.values()
}
```

```python
## 딕셔너리 표현식에서 if 조건문 사용하기 
x = {
    "a" : 10, 
    "b" : 20, 
    "c" : 30, 
    "d" : 40
}


for key, value in x.items():
    if value == 20:
        del x[key]

print(x)
```

```python
x = {
    "a" : 10, 
    "b" : 20, 
    "c" : 30, 
    "d" : 40
}

x = {key: value for key, value in x.items() if value != 20}
x 
```
