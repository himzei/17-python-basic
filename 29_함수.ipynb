```python
# 함수에서 값을 여러 개 반환하기 

def add_sub(a, b): 
    return a + b, a - b
```

```python
x, y = add_sub(10, 20)
print(x)
print(y)
```

```python
# 함수의 호출 과정 알아보기 

def mul(a, b):
    c = a * b
    return c

def add(a, b): 
    c = a + b
    print(c)
    d = mul(a, b)
    print(d)


x = 10
y = 20 
add(x, y)
```
